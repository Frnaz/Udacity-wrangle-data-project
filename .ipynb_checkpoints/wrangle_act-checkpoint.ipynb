```python
import pandas as pd
import requests
import os
import tweepy
import json
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Gather

```python
# loading twitter archive into a data frame
twitter_archive = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
# downloading image predictions
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
with open(os.path.join('image_predictions.tsv'), mode = 'wb') as file:
    file.write(response.content)
# loading image predictions into a data frame
image_predictions = pd.read_csv('image_predictions.tsv', sep = '\t')
```

```python
# accessing twitter API
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth_handler=auth,
                parser=tweepy.parsers.JSONParser(),
                wait_on_rate_limit=True,
                wait_on_rate_limit_notify=True)
```

```python
tweets = []
present_tweet_ids = []
missing_tweet_ids = []
for tweet_id in twitter_archive['tweet_id']:
    try:
        page = api.get_status(tweet_id, tweet_mode='extended')
        present_tweet_ids.append(tweet_id)
        tweets.append(page)
    except: # missing tweets are safed in a list
        missing_tweet_ids.append(tweet_id)
        print(tweet_id)
```

```python
len(missing_tweet_ids)
```

```python
with open('tweet_json.txt', mode = 'w') as file: # writing tweet data to a file
    json.dump(tweets, file)
```

```python
twitter_df = pd.read_json('tweet_json.txt')
twitter_df['tweet_id'] = present_tweet_ids
twitter_df = twitter_df[['tweet_id', 'favorite_count', 'retweet_count']]
```

## Clean

### Quality issues

```python
# copy existing data
twitter_archive_clean = twitter_archive.copy()
twitter_df_clean = twitter_df.copy()
image_predictions_clean = image_predictions.copy()
```

1. some tweets in twitter_archive are retweets

```python
# code
twitter_archive_clean = twitter_archive_clean[pd.isnull(twitter_archive_clean['retweeted_status_user_id'])]

# test
print(sum(twitter_archive_clean.retweeted_status_user_id.value_counts()))
```

2. some tweets are replies

```python
# code 
twitter_archive_clean = twitter_archive_clean[pd.isnull(twitter_archive_clean['in_reply_to_status_id'])]

# test 
print(sum(twitter_archive_clean.in_reply_to_status_id.value_counts()))
```

3. in twitter archive datetime is timestamp is type object

```python
# code
twitter_archive_clean['timestamp'] = pd.to_datetime(twitter_archive_clean['timestamp'])

# test
twitter_archive_clean.info()
```

4. some names in twitter_archive are wrong including
    * a
    * an
    * the

```python
# code
some_wrong_names = ['a', 'an', 'the']
twitter_archive_clean['name'].replace(some_wrong_names, 
                                np.nan,
                                inplace = True)

# test
twitter_archive_clean.name.value_counts()
```

5. in twitter archive there are some empty values saved as 'None' instead of NaN

```python
# code
columns = ['name', 'doggo', 'floofer', 'pupper', 'puppo']
for column in columns:
    twitter_archive_clean[column].replace('None', 
                                np.nan,
                                inplace = True)
# test
twitter_archive_clean.sample(7)
```

6. duplicated images in image predictions

```python
# code
image_predictions_clean = image_predictions_clean.drop_duplicates(subset=['jpg_url'])

# test
sum(image_predictions_clean.jpg_url.duplicated())
```

7. some rating numerators and denominators are off scale

```python
# code
# from studying the twitter page it seems unlikely that the base for the rating would be anything else than 10
twitter_archive_clean.drop(twitter_archive_clean[twitter_archive_clean.rating_denominator > 10].index, inplace=True)
twitter_archive_clean.drop(twitter_archive_clean[twitter_archive_clean.rating_denominator < 10].index, inplace=True)
# no dog should have a rating below 100% (why? because the're good dogs)
twitter_archive_clean.drop(twitter_archive_clean[twitter_archive_clean.rating_numerator < 10].index, inplace=True)

# test
print(twitter_archive_clean.rating_denominator.value_counts(), twitter_archive.rating_numerator.value_counts())
```

8. in image predictions p1, p2 and p3 have some lower letter words seperated by underscore

```python
# code
image_predictions_clean.p1.replace('_', ' ', regex=True, inplace=True)
image_predictions_clean.p1 = image_predictions_clean.p1.str.title()
image_predictions_clean.p2.replace('_', ' ', regex=True, inplace=True)
image_predictions_clean.p2 = image_predictions_clean.p2.str.title()
image_predictions_clean.p3.replace('_', ' ', regex=True, inplace=True)
image_predictions_clean.p3 = image_predictions_clean.p3.str.title()

# test
image_predictions_clean.head()
```

### Tidiness issues

category values (doggo, floofer, ...) in twitter_archive in four columns (should be one)

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.combine_first.html#pandas.DataFrame.combine_first
# code
twitter_archive_clean['stage'] = twitter_archive_clean['pupper'].combine_first(twitter_archive_clean['floofer']).combine_first(twitter_archive_clean['doggo'])
twitter_archive_clean['stage'] = twitter_archive_clean['stage'].astype('category')

# test
twitter_archive_clean.stage.value_counts()
```

three tables should be combined as they are about the same tweets

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
# code 
df = twitter_archive_clean.merge(twitter_df_clean, how='left', on='tweet_id')
df = df.merge(image_predictions_clean, how='left', on='tweet_id')

# test
df.columns
```

drop unneccessary columns

```python
# code
# rating denominator can be dropped because it's always 10
df.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'rating_denominator', 'retweeted_status_timestamp', 'doggo',
        'floofer', 'pupper', 'puppo'], axis=1, inplace=True)

# test 
df.columns
```

### Writing file

```python
df.to_csv('twitter_archive_master.csv', encoding='utf-8')
```
